```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print ("all libraries loaded sucessfuly")
```

```python
df = pd.read_csv('../data/raw/telco-Customer-Churn.csv')
print(f"Dataset shape: {df.shape}")
print(f"Total Customers: {df.shape[0]}")
print(f"Total Features: {df.shape[1]}")
```

```python
print("=== COLUMN INFO ===\n")
print(df.dtypes)
```

```python
churn_counts = df['Churn'].value_counts()
churn_pct = df['Churn'].value_counts(normalize=True) * 100

print("=== CHURN DISTRIBUTION ===")
print(f"Not Churned : {churn_counts['No']}  ({churn_pct['No']:.1f}%)")
print(f"Churned     : {churn_counts['Yes']} ({churn_pct['Yes']:.1f}%)")
print(f"\nThis means {churn_pct['Yes']:.1f}% of customers left the company.")
print(f"That is your core business problem.")
```

```python
print("=== MISSING VALUES ===\n")
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
})

print(missing_df[missing_df['Missing Count'] > 0])
print("\nColumns with zero missing values — clean ✓" if missing.sum() == 0 else "")
```

```python
print("=== NUMERICAL COLUMNS SUMMARY ===\n")
df.describe()
```

```python
duplicates = df['customerID'].duplicated().sum()
print(f"Duplicate Customer IDs: {duplicates}")

if duplicates == 0:
    print("No duplicates found — each row is a unique customer ✓")
else:
    print(f"WARNING: {duplicates} duplicate IDs found — needs cleaning!")
```

```python
print("=== UNIQUE VALUE COUNTS PER COLUMN ===\n")
for col in df.columns:
    print(f"{col:25s} → {df[col].nunique()} unique values")
```

```python
summary = {
    'total_rows': df.shape[0],
    'total_columns': df.shape[1],
    'churn_rate_%': round(churn_pct['Yes'], 2),
    'missing_values': int(df.isnull().sum().sum()),
    'duplicate_ids': int(duplicates)
}

print("=== PROJECT DATA PROFILE SUMMARY ===\n")
for key, value in summary.items():
    print(f"  {key:25s}: {value}")
```
