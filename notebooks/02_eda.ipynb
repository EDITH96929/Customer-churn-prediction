```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Plot styling — makes charts look professional
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['font.size'] = 12

df = pd.read_csv('../data/raw/telco-Customer-Churn.csv')
print(f"Data loaded: {df.shape}")
```

```python
# TotalCharges looks numerical but is actually stored as text
print(df['TotalCharges'].dtype)
print(df['TotalCharges'].sample(10))
```

```python
# Convert and find hidden missing values
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(f"\nHidden nulls after conversion: {df['TotalCharges'].isnull().sum()}")
# You'll find 11 hidden missing values that isnull() missed before
```

```python
# These 11 customers have 0 tenure — they just joined, fill with 0
df['TotalCharges'].fillna(0, inplace=True)
print("Fixed ✓")
print(f"Missing values now: {df['TotalCharges'].isnull().sum()}")
```
